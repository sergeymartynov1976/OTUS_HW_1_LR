```scala
import $ivy.`com.github.tototoshi::scala-csv:1.3.6`
import com.github.tototoshi.csv._
import java.io._
import $ivy.`org.scalanlp::breeze:1.0`
import breeze.linalg._
import breeze.numerics._
import breeze.stats.regression._
```

#### Creat necessary methods

```scala
// Reader of csv
def read(name: String): List[List[String]] = {
    val reader = CSVReader.open(new File(name))
    reader.all()
}
```

```scala
// Preprocessor to work with features
def f_preproc(data: List[List[String]]): DenseMatrix[Double] = {
    val f = data.slice(1,data.length).map(x=> x.toArray) // Take off header row
    val f1 = f.map(_.map(_.toDoubleOption.getOrElse(0.0))) // Convert features into Double type
    DenseMatrix(f1:_*) // Convert data into Dence Matrix
}
```

```scala
// Predictors creater
def predictors(features: DenseMatrix[Double]): DenseMatrix[Double] = {
    val ones = DenseVector.ones[Double](features(::,0).length)
    DenseMatrix.horzcat( ones.asDenseMatrix.t, features(::, 0 to 26))
}
```

```scala
// Targets creator
def targets(features: DenseMatrix[Double]): DenseVector[Double] = {
    features(::, 27)
}
```

```scala
//RMSE calculator
def rmse(predictions: DenseVector[Double], targets: DenseVector[Double]): Double = {
    sqrt(sum((predictions - targets)*:*(predictions - targets))/predictions.length)
}
```

```scala
//Breeze Linear-Regression model
def breeze_lin_regr(predictors: DenseMatrix[Double], targets: DenseVector[Double]): DenseVector[Double] = {
    val model = leastSquares(predictors, targets)
    model.coefficients
}
```

```scala
//Breeze Lasso-Regression model
def breeze_lasso_regr(predictors: DenseMatrix[Double], targets: DenseVector[Double]): DenseVector[Double] = {
    val model = lasso(predictors, targets, 0.01)
    model.coefficients
}
```

```scala
//My Linear-Regression model through normal equation
def my_lin_regr(predictors: DenseMatrix[Double], targets: DenseVector[Double]): DenseVector[Double] = {
   inv(predictors.t*predictors)*predictors.t*targets
}
```

#### Read file and process the data

```scala
val name = "Dataset.csv"
```

```scala
val predictors1 = predictors(f_preproc(read(name)))
```

```scala
var targets1 = targets(f_preproc(read(name)))
```

#### Calculate prediction and define RMSE

```scala
// Prediction with breeze lin regr model
val prediction_blinreg = predictors1*breeze_lin_regr(predictors1, targets1)
```

```scala
// Prediction with breeze lasso regr model
val prediction_blassoreg = predictors1*breeze_lasso_regr(predictors1, targets1)
```

```scala
// Prediction with my linear regr model
val prediction_my_linreg = predictors1*my_lin_regr(predictors1, targets1)
```

##### Calculating RMSEs

```scala
val blinregr_rmse = rmse(prediction_blinreg, targets1)
val blassoregr_rmse = rmse(prediction_blassoreg, targets1)
val mylinregr_rmse = rmse(prediction_my_linreg, targets1)
println("RMSE of breeze linear regression model:" + blinregr_rmse)
println("RMSE of breeze lasso regression model:" + blassoregr_rmse)
println("RMSE of my linear regression model:" + mylinregr_rmse)
```
